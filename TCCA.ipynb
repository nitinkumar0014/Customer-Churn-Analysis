```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\nitin\Downloads\Customer Churn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#Replacing Blanks with 0 as tenure is 0
```

```python
df["TotalCharges"]= df["TotalCharges"].replace(" ","0")
df["TotalCharges"]= df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
ax=sns.countplot(x='Churn',data = df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Customer By Churn")
plt.show()
```

```python
#from the given pie chart we can conclude that 26.54% of our customer churned out
# let's find the reason behind it
```

```python
ax = sns.countplot(x='SeniorCitizen', data=df, hue='Churn')

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
#“The churned customer count is higher among senior citizens compared to younger customers .”
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x="tenure", data =df,bins = 72, hue ="Churn")
plt.show()
```

```python
# people who used our services for a long time have stayed and people who have used services 1 or 2 months have churned out
```

```python
ax = sns.countplot(x='Contract', data=df, hue='Churn')

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
#people who have month to month contract are likely churned out as compared to long term contract
```

```python
df.columns.values
```

```python
services = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(3, 3, figsize=(18, 15))

for ax, col in zip(axes.flatten(), services):
    
    sns.countplot(x=col, data=df, hue='Churn', ax=ax)
    
    ax.set_title(f'{col} vs Churn')
    ax.tick_params(axis='x', rotation=20)
    
    # Add labels on bars
    for container in ax.containers:
        ax.bar_label(container)

plt.tight_layout()
plt.show()
```

```python
# Customers without OnlineSecurity and TechSupport services, especially Fiber optic users, show the highest churn rate.
```

```python
ax = sns.countplot(x='PaymentMethod', data=df, hue='Churn')

for container in ax.containers:
    ax.bar_label(container)
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

```python
# The highest chunk rate in customers due to electronic check payment method
```
